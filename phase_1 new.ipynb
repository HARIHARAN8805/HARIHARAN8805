## Handling Outliers

```python
sns.boxplot(y='avg_glucose_level',data=df)
```

```python
q1=df.avg_glucose_level.quantile(0.25)  
q3=df.avg_glucose_level.quantile(0.75)
```

```python
IQR=q3-q1
```

```python
upper_limit= q3 + 1.5*IQR

lower_limit= q1 - 1.5*IQR
```

```python
upper_limit
```

```python
lower_limit
```

```python
df.median()
```

```python
df['avg_glucose_level']= np.where(df['avg_glucose_level']>upper_limit,91.85,df['avg_glucose_level'])
```

```python
sns.boxplot(y='avg_glucose_level',data=df)
```

```python
df["avg_glucose_level_capped"] = np.where(df["avg_glucose_level"]> upper_limit, upper_limit,
                        np.where(df["avg_glucose_level"]< lower_limit, lower_limit,
                          df["avg_glucose_level"]))
```

```python
sns.boxplot( y="avg_glucose_level_capped", data=df)
```

## One Hot Encoding

## X and Y split

```python
# dependent variable
y=df_1['stroke']
y
```

```python
#independent variable

X=df_1.drop(columns=['stroke'],axis=1)
X.head(10)
```

## Scaling

```python
from sklearn.preprocessing import scale
```

```python
x_scaled=pd.DataFrame(scale(X),columns=X.columns)
x_scaled.head()
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x_scaled,y,test_size=0.3,random_state=3)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix
from sklearn import metrics
```

## Logistic Regression 

```python
log_reg=LogisticRegression(random_state=3)
log_reg.fit(X_train,y_train)
y_pred=log_reg.predict(X_test)
```

```python
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

```python
print('Training-set accuracy score:', log_reg.score(X_train, y_train))
print('Test-set accuracy score:', log_reg.score(X_test, y_test))
```

```python
acc_log_clf = metrics.accuracy_score(y_test,y_pred)
acc_log_clf
```

```python
pred_list = [log_reg]

for i in pred_list:
    print("Score : ",i.score(X_test,y_test))
    y_pred = i.predict(X_test)
    sns.heatmap(confusion_matrix(y_test,y_pred),annot = True)
    plt.xlabel("Y_pred")
    plt.ylabel("Y_test")
    plt.title(i)
    plt.show()
```
